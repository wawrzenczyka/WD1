```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = pd.read_csv('../data/budgets.csv')
df['id'] = pd.to_numeric(df['name'].str[9:])
df = df.drop(columns=['name'])
#df = df.set_index('id')
df.head()
```

```python
fig, ax = plt.subplots() 
fig.set_size_inches(8,4)
g = sns.barplot(x='id', y='budget', data=df, ax=ax, palette="Blues_d")
ax.set(xlabel='edycja PRELUDIUM')
ax.set(ylabel='budÅ¼et (mln)')
plt.show()
```

```python
fig, ax = plt.subplots() 
fig.set_size_inches(8,4)
g = sns.lineplot(x='id', y='budget', data=df, ax=ax)
ax.set(xlabel='edycja PRELUDIUM')
ax.set(ylabel='budÅ¼et (mln)')
plt.show()
```
